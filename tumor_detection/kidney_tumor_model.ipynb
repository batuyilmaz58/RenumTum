```python
!pip install supervision -q
import supervision as sv
import cv2
from ultralytics import YOLO  # YOLO import'unu ekledim

model = YOLO("/workspace/runs/detect/train3/weights/best.pt")

# GÃ¶rÃ¼ntÃ¼yÃ¼ cv2 ile yÃ¼kleyin
image = cv2.imread("/workspace/Kidney-Stone-Detection-1/test/images/1-3-46-670589-33-1-63740171276284352500001-5536100503549537984_png_jpg.rf.eee63637eedbcadd53a04531692326a8.jpg")

# Class mapping - class_id ile mapping yapÄ±n, string ile deÄŸil
class_name_mapping = {
    0: "BÃ¶brek TaÅŸÄ±"
}

# Detection sonuÃ§larÄ±nÄ±zÄ± aldÄ±ktan sonra
results = model(image)[0]

# Ultralytics YOLO sonuÃ§larÄ±nÄ± Supervision format'Ä±na dÃ¶nÃ¼ÅŸtÃ¼rÃ¼n
detections = sv.Detections.from_ultralytics(results)

# Supervision ile gÃ¶rselleÅŸtirme
box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

# Ã–zel etiketler oluÅŸturma - class_id kullanarak mapping yapÄ±n
labels = []
for class_id in detections.class_id:
    display_name = class_name_mapping.get(class_id, f"Class_{class_id}")
    confidence = detections.confidence[len(labels)] if detections.confidence is not None else 0.0
    labels.append(f"{display_name} {confidence:.2f}")

# Annotation
annotated_image = box_annotator.annotate(image, detections)
annotated_image = label_annotator.annotate(annotated_image, detections, labels=labels)

# Sonucu gÃ¶ster
cv2.imshow('Detection Result', annotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Veya kaydet
cv2.imwrite('result.jpg', annotated_image)
```
